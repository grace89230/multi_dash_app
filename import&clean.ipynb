```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dairy_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Production Date'] = pd.to_datetime(df['Production Date'])
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'])
```

```python
df.isnull().sum()
```

```python
# Convert the date column to datetime and sort in ascending order
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.describe().T
```

```python
df.columns
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Shelf Life (days)'], df['Quantity in Stock (liters/kg)'], marker='o', linestyle='-', color = 'k')
plt.title('Shelf Life and Quantity in Stock')
plt.xlabel('Shelf Life')
plt.ylabel('Quantity Sold (liters/kg)')
plt.grid(True)
plt.show()

```

```python
# Assuming 'Sales Channel' contains sales channels
sales_channel_distribution = df['Customer Location'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
sales_channel_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Product Distribution')
plt.ylabel('')
plt.show()
```

```python
# Assuming 'Customer Location' contains customer locations
customer_distribution = df['Customer Location'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
customer_distribution.plot(kind='bar', color='m')
plt.title('Customer Distribution')
plt.xlabel('Customer Location')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
